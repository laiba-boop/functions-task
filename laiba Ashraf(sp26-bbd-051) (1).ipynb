```python
#question no 2
def calculate_average(a, b, c):
    return (a + b + c) / 3

print("Average:", calculate_average(10, 20, 30))

```

```python
#question no 3

def calculate_discount(price, discount=10):
    final_price = price - (price * discount / 100)
    return final_price

print("Discounted Price:", calculate_discount(1000))
print("Discounted Price with 20%:", calculate_discount(1000, 20))
```

```python
#question no 5
def find_max(*numbers):
    return max(numbers)

print("Maximum:", find_max(3, 7, 2, 9, 5))
```

```python
#question no 6
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial:", factorial(5))
```

```python
#question no 7

def apply_function(func, numbers):
    return [func(num) for num in numbers]

def double(x):
    return x * 2

print(apply_function(double, [1, 2, 3, 4]))

```

```python
#question no 9
def apply_operation(operation, numbers):
    return [operation(num) for num in numbers]

print(apply_operation(lambda x: x + 10, [1, 2, 3, 4]))
```

```python
#question no 10

import time

def timer_decorator(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f"Execution Time: {end - start:.6f} seconds")
    return wrapper

@timer_decorator
def sample_function():
    time.sleep(1)
    print("Function is running...")

sample_function()
```
